```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from rasterio.plot import show

with rasterio.open("subset_dem.tif") as src:

    fig, ax = plt.subplots(figsize=(8, 8))

    img = show(
        src,
        ax=ax,
        cmap="terrain"
    )

    cbar = plt.colorbar(
        img.get_images()[0],
        ax=ax
    )

    cbar.set_label("Elevation (m)")
    ax.set_title("Subset DEM")

    plt.show()
```
